### Portfolio Application

**1. CARA utility function: **
$$
U ( x ) = \frac { - e ^ { - a x } } { a } \text { for } a \neq 0
$$
$A ( x ) = \frac { - U ^ { \prime \prime } ( x ) } { U ^ { \prime } ( x ) } = a$ is Coefficient of Constant Absolute Risk-Aversion.

If the random outcome $x \sim \mathcal { N } \left( \mu , \sigma ^ { 2 } \right)$,
$$
\mathbb { E } [ U ( x ) ] = \left\{ \begin{array} { l l } { \frac { - e ^ { - s \mu + \frac { a ^ { 2 } \sigma ^ { 2 } } { 2 } } } { a } } & { \text { for } a \neq 0 } \\ { \mu } & { \text { for } a = 0 } \end{array} \right.
$$
$$
x _ { C E } = \mu - \frac { a \sigma ^ { 2 } } { 2 }
$$

Portfolio Wealth is modelled as 
$$
W \sim \mathcal { N } \left( 1 + r + \pi ( \mu - r ) , \pi ^ { 2 } \sigma ^ { 2 } \right)
$$
From the section on CARA Utility, we know we need to maximize:
$$
1 + r + \pi ( \mu - r ) - \frac { a \pi ^ { 2 } \sigma ^ { 2 } } { 2 }
$$
So optimal investment fraction in risky asset
$$
\pi ^ { * } = \frac { \mu - r } { a \sigma ^ { 2 } }
$$
